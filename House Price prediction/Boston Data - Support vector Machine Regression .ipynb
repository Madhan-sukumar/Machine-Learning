## Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.svm import SVR
```

## Loading boston dataset from sklearn datasets

```python
from sklearn.datasets import load_boston
```

```python
boston_data = load_boston()
```

```python
boston_data
```

```python
boston_data.feature_names
```

### dataset got two sections, so combining both section before converting into dataframe

```python
data = pd.DataFrame(boston_data.data,columns=boston_data.feature_names)
```

```python
data.head(10)
```

```python
data['MEDV']=boston_data.target
```

```python
data.head(10)
```

```python
data.describe()
```

```python
data.info()
```

## Data Preprocessing

```python
#checking any null values is present
data.isnull().sum()
```

## Data splitting

```python
x = data.drop('MEDV',axis=1)
```

```python
y = data['MEDV']
```

```python
x.head(10)
```

```python
y.head(10)
```

## Data Scaling

```python
scaler = StandardScaler()
```

```python
x =scaler.fit_transform(x)
```

```python
x
```

## Dataset splitting for training & testing

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.3, random_state = 0)
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
ytrain.shape
```

```python
ytest.shape
```

## Support Vector Regression

```python
#by trail & error , the best kernel fit for this model is rbf. The Error percentage using of all kernels are
#linear = 38.84 %
#rbf = 36.99 %
#poly = 45.093 %
#sigmoid = 53.05 % 

svr = SVR(kernel='rbf')
```

```python
svr.fit(xtrain,ytrain)
```

## Predict the output

```python
predicted_value = svr.predict(xtest)
```

```python
predicted_value 
```

```python
ytest
```

## calculate the error between predicted output and test output

```python
error = mean_squared_error(predicted_value ,ytest)
```
